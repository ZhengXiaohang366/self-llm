```python
import json
import re

# 方法一：使用 with 语句自动管理文件的打开和关闭
file_path = '../dataset/res/test_json1.json'
with open(file_path, 'r', encoding='utf-8') as file:
    # 读取并解析 JSON 数据
    test_data1 = json.load(file)

file_path = '../dataset/res/test_json2.json'
with open(file_path, 'r', encoding='utf-8') as file:
    # 读取并解析 JSON 数据
    test_data2 = json.load(file)

file_path = '../dataset/res/test_json3.json'
with open(file_path, 'r', encoding='utf-8') as file:
    # 读取并解析 JSON 数据
    test_data3 = json.load(file)

file_path = '../dataset/res/test_json4.json'
with open(file_path, 'r', encoding='utf-8') as file:
    # 读取并解析 JSON 数据
    test_data4 = json.load(file)
```

```python
test_data = test_data1[:50] + test_data2[50:100] + test_data3[100:150] + test_data4[150:200] 
```

```python
res = []
for d in test_data:
    tmp_dict={}
    tmp_dict["ID"] = d['ID']
    
    text =d['output']
    # 使用正则表达式匹配处方部分
    pattern = r"'处方':'(\[.*?\])'"
    match = re.search(pattern, text)
    if match:
        prescription_str = match.group(1)
    else:
        print(d['ID']+"未找到处方信息。")
    tmp_dict["子任务2"] = prescription_str
    
    input_str =d['output']
    input_str = input_str.replace("'", '"')
    # 处理处方字段里嵌套列表的引号问题
    input_str = input_str.replace('"[', '[').replace(']"', ']')
    
    try:
        # 使用 json.loads 方法将字符串解析为字典
        data_dic = json.loads(input_str)
    except:
        print("失败。")
    tmp_dict["子任务1"] = [data_dic['证型'],data_dic['疾病']]
    res.append(tmp_dict)
```

```python
# 定义保存文件的路径
file_path = '../dataset/res/final_res.json'

# 打开文件并写入数据
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(res, f, ensure_ascii=False, indent=4)

print(f"数据已保存到 {file_path}")
```
