```python
import json

# 方法一：使用 with 语句自动管理文件的打开和关闭
file_path = '../dataset/res/test_json.json'
with open(file_path, 'r', encoding='utf-8') as file:
    # 读取并解析 JSON 数据
    test_data = json.load(file)
    # 打印读取的数据
    print(test_data[0])
```

```python
res = []
for d in  test_data:
    tmp_dic = {}
    tmp_dic["病案标识"] = d['ID']
    tmp_dic["预测结果"] = d['output'].replace('{\'主要诊断编码\':\'','["').replace('\', \'其他诊断编码\':\'','"|"').replace('\'}','"]')
    res.append(tmp_dic)
```

```python
res
```

```python
# 定义保存文件的路径
file_path = '../dataset/res/final_res.json'

# 打开文件并写入数据
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(res, f, ensure_ascii=False, indent=4)

print(f"数据已保存到 {file_path}")
```
